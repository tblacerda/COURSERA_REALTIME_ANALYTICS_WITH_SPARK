```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("QuickCommerce Streaming Pipeline") \
    .config("spark.jars.packages", "org.apache.spark:spark-sql-kafka-0-10_2.12:3.3.1") \
    .getOrCreate()
    
```

```python
# use the SparkSession `spark` created in a previous cell
# read from Kafka and cast key/value to strings for easier downstream processing.
# also disable failing on data loss to avoid errors when offsets are unavailable in local/test environments.
streaming_df = spark.readStream \
	.format("kafka") \
	.option("kafka.bootstrap.servers", "localhost:9092") \
	.option("subscribe", "ecommerce_topic") \
	.option("startingOffsets", "earliest") \
	.option("failOnDataLoss", "false") \
	.load() \
	.selectExpr(
		"CAST(key AS STRING) AS key",
		"CAST(value AS STRING) AS value",
		"topic",
		"partition",
		"offset",
		"timestamp"
	)
```
