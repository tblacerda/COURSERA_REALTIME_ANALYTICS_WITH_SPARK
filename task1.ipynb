### TASK 1
-----

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("QuickCommerce Streaming Pipeline") \
    .config(
        "spark.jars.packages",
        "org.apache.spark:spark-sql-kafka-0-10_2.13:4.0.1"
    ) \
    .getOrCreate()
```

```python
# use the SparkSession `spark` created in a previous cell
streaming_df = spark.readStream \
	.format("kafka") \
	.option("kafka.bootstrap.servers", "localhost:9092") \
	.option("subscribe", "ecommerce_topic") \
	.option("startingOffsets", "earliest") \
	.load()
```

```python
checkpoint_dir = "/tmp/quickcommerce_streaming_checkpoint"
```

```python
query = streaming_df.writeStream \
.format("console").outputMode("append").option("checkpointLocation", checkpoint_dir).start()
```

```python
streaming_df.printSchema()
```

### TASK 2
----

```python
streaming_df = streaming_df.selectExpr("CAST(value AS STRING) as json")\
    .selectExpr("from_json(json, 'transaction_id STRING, user_id STRING, item_id STRING, amount DOUBLE, timestamp STRING') AS data")\
    .select("data.*")
```

```python
streaming_query = streaming_df.writeStream \
.trigger(processingTime='10 seconds') \
.option("checkpointLocation", checkpoint_dir)\
.format("console") \
.start()
```

## TASK 3
---

```python
from pyspark.sql.functions import to_timestamp

streaming_df = streaming_df.withColumn(
    "timestamp",
    to_timestamp("timestamp", "yyyy-MM-dd HH:mm:ss")
)

streaming_df = streaming_df.withWatermark(
    "timestamp",
    "10 minutes"
)

filtered_df = streaming_df.filter("amount > 500")
```

```python
from pyspark.sql.functions import sum, window

aggregated_df = filtered_df.groupBy(
    window(filtered_df.timestamp, "10 minutes", "5 minutes"),
    filtered_df.user_id
).agg(
    sum("amount").alias("total_amount")
)
```

```python
query = aggregated_df.writeStream \
    .format("json") \
    .option("path", "/tmp/output/aggregated_checkpoint") \
    .option("checkpointLocation", "/tmp/aggregation_checkpoint") \
    .start()
```
